```python
data = pd.read_csv("/content/Fuel_cell_performance_data-Full.csv")
```

```python
df = pd.DataFrame(data)

df_target3 = df[['F1', 'F2', 'F3', 'F4', 'F5', 'F6', 'F7', 'F8', 'F9', 'F10', 'F11', 'F12', 'F13', 'F14', 'F15', 'Target3']]
```

```python
df_target3
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df_target3, test_size=0.3, random_state=42)
```

```python
X_train = train.drop('Target3', axis=1)
y_train = train['Target3']
X_test = test.drop('Target3', axis=1)
y_test = test['Target3']
```

```python
regression_setup = setup(data=train, target='Target3', session_id=42, normalize=True)
```

```python
best_model = compare_models()
```

```python
evaluate_model(best_model)
```

```python
final_model = finalize_model(best_model)
```

```python
predictions = predict_model(final_model, data=X_test)
```
